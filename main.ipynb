```python
import investpy as inv
br = inv.stocks.get_stocks(country='brazil')
```

```python
carteira = []
for a in br['symbol']:
  if len(a) <= 5:
    carteira.append(a+'.SA')
```

```python
dt = yf.download(carteira, start = '2015-01-01', end = '2024-01-01')['Adj Close']
```

```python
"""novaTabela= []
for i in range(len(dt)):
  for j in range(len(dt)):
    novaTabela = (dt[i][j] - dt[i][j-1])/ dt[i][j-1]"""

nova_tabela = []
for i in range(len(dt)):
  nova_linha = []  # Create a new list for each row's differences
  for j in range(1, len(dt[i])):  # Start from j=1 (skip the first element)
    diferenca = (dt[i][j] - dt[i][j-1]) / dt[i][j-1]
    nova_linha.append(diferenca)
  nova_tabela.append(nova_linha)  # Add the row of differences to nova_tabela

# (Optional) Convert nova_tabela to a pandas DataFrame for further manipulation
df = pd.DataFrame(nova_tabela)
```
